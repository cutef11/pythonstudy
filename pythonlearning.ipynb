```python

def isvalid(s):
    dic={'）':'（',']':'[','}':'{'}
    stack=[]
    
    for ch in s:
        if ch in ("([{"):
            stack.append(ch)
        else:
            if not stack or (dic[ch]!=stack[-1]):
                return False
            stack.pop()
    return not stack    


isvalid("[[]]")
```

```python
class ListNode:
    def __init__(self, val=0, next=None):  # 注意是 __init__
        self.val = val      # 当前节点的值
        self.next = next    # 指向下一个节点

def printlist(head):
    cur = head
    while cur:
        print(cur.val)
        cur = cur.next


def buildlist(a):
    if not a:
        return None

    head = ListNode(a[0])  # 先建头结点
    cur = head

    for i in range(1, len(a)):
        newnode = ListNode(a[i])
        cur.next = newnode
        cur = cur.next

    return head

def reverselist(head):
  
    pre=None
    curr=head
    while(curr):
            
        back=curr.next
            
        curr.next=pre
            
        pre=curr
        curr=back
    return pre

bb = buildlist([1, 2, 3, 5, 6, 9])
printlist(bb)

dd = buildlist([1, 2, 3, 5, 6, 9])
printlist(dd)


def mergelist(a,b):
    dummy=ListNode(0)
    curr=dummy
    
    while(a and b):
        if(a.val>b.val):
            curr.next=b
            b=b.next
            
        else:
            curr.next=a
            a=a.next
        curr=curr.next
    
    curr.next = a if a else b

    return dummy.next


ee=mergelist(bb,dd)
printlist(ee)
```
